```python
import openml
import pandas as pd
```

```python
def getValidDatasets(numFeatures = 50, numInstances=1000, numClasses=15,
                         missingValues = False, numericFeaturesOnly = True):
    openml_list = openml.datasets.list_datasets()  # returns a dict

    # Show a nice table with some key data properties
    datalist = pd.DataFrame.from_dict(openml_list, orient="index")

    datalist = datalist[datalist["NumberOfFeatures"]< numFeatures]
    datalist = datalist[datalist["NumberOfInstances"]< numInstances]
    datalist = datalist[datalist["NumberOfClasses"]< numClasses]

    if not missingValues:
        datalist = datalist[datalist["NumberOfMissingValues"]== 0]
    if numericFeaturesOnly:
        datalist= datalist[datalist["NumberOfSymbolicFeatures"]==0]

    return datalist
    
    

```

```python
def describeDataset(dataset):
    propertiesList=dataset.qualities
    propertiesDf = pd.DataFrame(propertiesList.items(), columns=["key", "value"])
    AUCList = propertiesDf[propertiesDf["key"].str.contains("AUC")]
    meanAUC = AUCList.mean()
    
    return meanAUC
        
    
```

```python
validDatasets= getValidDatasets()
validDatasetsId = validDatasets.index.tolist()
datasetsArray = []
for id in validDatasetsId:
    dataset = openml.datasets.get_dataset(id, download_data=False)
    mean = describeDataset(dataset)
    print(id, mean)
    datasetsArray.append([id, mean])
```
